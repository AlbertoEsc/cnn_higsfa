```python
import numpy as np
#     0|2|4              model|data partition
#                           v       v       
#      1|3                  samples
#                               v 
total_successes = np.zeros((3,  7,  2))
total_lengths =   np.zeros((3,  7,  2))
```

```python
import glob
model = "cnn_2"
model_root = "experimental_data_mnist" + "/" + model
id0 = 0
print(sorted(glob.glob(model_root + "/*")))
print("\n")
for id1, samplesno in enumerate(sorted(glob.glob(model_root + "/*"))):    
    for id2, data_partition in enumerate(sorted(glob.glob(samplesno + "/*"))):    
        print(data_partition)
        total_success = 0
        total_length = 0     
        pct = []
        for id3, runno in enumerate(sorted(glob.glob(data_partition + "/*"))):
            with open(runno) as f:
                content = f.readlines()
                success = 0                
                for line in content:
                    split = line.split(",")
                    if split[1].strip() == split[2].strip():
                        success = success + 1
                total_success +=  success
                total_length  += len(content)-1
                pct.append(success/float(len(content)-1))
        print(np.std(pct) / np.sqrt(10.0))
        if id2 == 1:
            #print("total training samples: total_length / 100 = %i" %(total_length/100))
            print("classification error: %f" % (total_success / float(total_length)) )
        else:
            #print("test samples per iteration: %i" %(total_length/100))
            print("classification error: %f" % (total_success / float(total_length)) )
        print("ids %i, %i, %i " % (id0, id1, id2))
        total_successes[id0, id1, id2] += total_success
        total_lengths  [id0, id1, id2] += total_length
        print("\n")
                    
```
